## Introduction

## Data Preprocessing and Analysis

```python
print('There are', winners['Player ID'].nunique(), 'unique lottery winners in 2024.')
```

```python
winners['Player ID'] = winners['Player ID'].astype(str) # The player IDs would show up in scientific notation form because the full player ID was a huge number
anon = winners[winners['Anonymous Indicator'] == 'Yes']
print('There are', anon['Player ID'].nunique(), 'unique anonymous lottery winners in 2024.')
```

```python
winners['Amount Won'] = winners['Amount Won'].replace('[\$,]', '', regex=True)
winners['Amount Won'] = pd.to_numeric(winners['Amount Won'], errors='coerce')

winners['Scratch Ticket Cost'] = winners['Scratch Ticket Cost'].replace('[\$,]', '', regex=True)
winners['Scratch Ticket Cost'] = pd.to_numeric(winners['Scratch Ticket Cost'], errors='coerce')
```

```python
winners.dtypes
```

```python
winners.isna().sum()
```

```python
pd.set_option('display.float_format', '{:,.0f}'.format)
winners.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
game_counts = winners['Scratch Ticket or Draw Game'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=game_counts.index, y=game_counts.values, color="pink")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Game Type")
plt.ylabel("Number of Winning Claims")
plt.title("Game Type Distribution")
plt.tight_layout()
plt.show()
```

The most popular game type is Scratch Tickets.

```python
plt.hist(winners[winners['Amount Won'] <= 10000]['Amount Won'], bins=50, color='pink', edgecolor='black')
plt.xlabel("Amount Won ($)")
plt.ylabel("Number of Winning Claims")
plt.title("Distribution of Texas Lottery Prizes (≤ $10,000)")
plt.show()
```

This distribution of prizes makes sense because you're more likely to win a smaller prize than a larger prize at the lottery.

```python
winners['Claimant County'].value_counts().head(5)
```

The top five counties with the most lottery winners are: Harris, Bexar, Dallas, Tarrant, and Travis.

```python
win = winners # win = cleaned version of winners
```

```python
win = win.drop(['Last Name','First Name','Claimant Name and Player ID','Claim Center Location','Claimant Country','Retailer Number and Name','Prize Level','Scratch Ticket Game Number','Retailer Address 1','Retailer Address 2','Retailer Zip Code','Retailer Zip Code +4','Retailer City','Draw Time of Day'], axis=1)
```

```python
win = win.dropna(subset = ['Retailer County'])
```

```python
win.head()
```

```python
win['Date Claim Paid'] = pd.to_datetime(win['Date Claim Paid'], errors='coerce')
```

```python
win['Day of Week'] = win['Date Claim Paid'].dt.day_name()
win['Week of Month'] = win['Date Claim Paid'].apply(lambda d: (d.day - 1) // 7 + 1) # New Columns
```

```python
countyfips = pd.read_csv('Texas FIPS.csv', sep=',') # I got the FIPS codes from this website (https://www.gigasheet.com/sample-data/spreadsheet-list-of-all-counties-in-texascsv) and I deleted the unnecessary columns.
countyfips = countyfips.dropna(subset=['FIPS County']) # The Texas lottery dataset didn't have the FIPS codes in a column, so I had to join it with another dataset
countyfips['FIPS'] = countyfips['FIPS County'].astype(int).astype(str).str.zfill(3)
countyfips['FIPS'] = '48' + countyfips['FIPS']
countyfips.head() # Make sure the CSV file is working
```

```python
winnerstx = win[win['Claimant State'].isin(['TX', 'Texas'])].copy()
```

```python
winnerstx['Claimant County'] = winnerstx['Claimant County'].str.strip()
countyfips['County Name'] = countyfips['County Name'].str.strip()

countycorrections = { # I know to make these corrections because when I ran the code 2 cells below, it returned these as unmatched counties so I had to adapt
    'Mclennan': 'McLennan',
    'Mcmullen': 'McMullen',
    'Mcculloch': 'McCulloch',
    'Tom Green County': 'Tom Green',
    'Bexar County': 'Bexar',
    'Collin County': 'Collin',
    'Wise County': 'Wise'
}

winnerstx['Claimant County'] = winnerstx['Claimant County'].replace(countycorrections)
validcounties = countyfips['County Name'].tolist()
winnerstx = winnerstx[winnerstx['Claimant County'].isin(validcounties)]
```

```python
winnerstx = winnerstx.merge(countyfips[['County Name', 'FIPS']], left_on='Claimant County', right_on='County Name', how='left')
print(winnerstx[winnerstx['FIPS'].isna()]['Claimant County'].unique()) # If the above cells were done right, it should print an empty array
```

###**Research Question #1:** How are Texas lottery winners distributed across counties?

```python
import plotly.express as px

county_winners = winnerstx.groupby(['County Name', 'FIPS']).size().reset_index(name='Num Winners')
fig = px.choropleth(
    county_winners,
    geojson='https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json',
    locations='FIPS', color='Num Winners', color_continuous_scale="greens",
    scope="usa", labels={'Num Winners':'Number of Winners'},
    hover_data={'County Name': True, 'Num Winners': True, 'FIPS': False}
)
fig.update_geos(visible=False, scope="usa", fitbounds="locations")

fig.update_layout(
    title_text='Texas Lottery Winners by County',
    title_x=0.47,  # I wanted to center the title
    coloraxis_colorbar=dict(title="Number of Winners"),
    font=dict(family="Calibri", size=16, color="black")
)

fig.show()
```

The choropleth map shows the geographic distribution of Texas lottery winners by county. Darker green shades indicate counties with more winners, with Harris, Bexar, and Dallas counties having the highest counts, which makes sense because these counties are more populated.

###**Research Question #2:** How do the number of Texas lottery winners vary over time in 2024?

```python
daily_winners = win.groupby('Date Claim Paid').size().reset_index(name='Number of Winners')

fig = px.line(daily_winners, x='Date Claim Paid', y='Number of Winners', title='Texas Lottery Winners in 2024')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=7, label="1W", step="day", stepmode="backward"),
            dict(count=1, label="1M", step="month", stepmode="backward"),
            dict(count=3, label="3M", step="month", stepmode="backward"),
            dict(count=6, label="6M", step="month", stepmode="backward"),
            dict(step="all", label="All 2024")
        ])
    )
)

fig.update_traces(line=dict(color='green'))
fig.update_layout(font=dict(family="Calibri", size=16, color="black"))
fig.show()
```

The time series shows the number of Texas lottery winners recorded each day throughout 2024. Peaks indicate days with more claims, such as February 20, April 29, October 7, November 18, and December 27, likely due to scheduled lottery draws, promotional deadlines, or batch processing of claims. Valleys indicate days with fewer claims, including notable drops on January 15, July 4, and December 15, likely due to holidays (MLK Jr. Day, Independence Day) or office closures. Overall, the chart reveals temporal patterns in when winners claim their prizes.

### **Research Question #3:** How does the prize distribution vary for scratch tickets of different prices in Texas?

```python
scratch_winners = win[win['Scratch Ticket or Draw Game'] == 'Scratch Tickets'].copy()
scratch_winners = scratch_winners.dropna(subset=['Scratch Ticket Cost', 'Amount Won'])
scratch_winners['Ticket Price'] = '$' + scratch_winners['Scratch Ticket Cost'].astype(int).astype(str) # Conversion so the labels aren't messed up

fig = px.box(
    scratch_winners, y='Ticket Price', x='Amount Won', title='Prize Distribution by Scratch Ticket Price in Texas (2024)',
    labels={'Ticket Price': 'Scratch Ticket Price','Amount Won': 'Prize Amount Won ($)'}, points='outliers', orientation='h',
    category_orders={'Ticket Price': sorted(scratch_winners['Ticket Price'].unique(),key=lambda x: float(x.replace('$', '')))}
)

fig.update_traces(boxmean='sd', marker=dict(size=2, opacity=0.5, color='darkgreen'), fillcolor='darkgreen', line=dict(color='darkgreen'))

fig.update_layout(
    xaxis=dict(type='log', title='Prize Amount Won ($, log scale)', gridcolor='lightgray', showgrid=True), # The graph is in logarithmic scale because of the extreme outliers
    yaxis=dict(title='Scratch Ticket Price', showgrid=False), plot_bgcolor='white', paper_bgcolor='white', font=dict(family="Calibri", size=16, color="black"),
    showlegend=False, height=500, margin=dict(l=120, r=80, t=80, b=80)
)

fig.show()
```

The box plot figure shows the distribution of prize amounts for scratch tickets of different prices in Texas. Each horizontal box represents a ticket price, and individual points indicate outliers. Higher-priced tickets tend to have larger prizes and more variability, including rare extreme wins, showing how prize amounts vary by ticket price.

### **Research Question #4:** Which retailers in Texas have the most lottery winners, and how does this vary by game type or county?

```python
import plotly.graph_objects as go

retailerwinners = win.groupby(['Retailer Name', 'Retailer County', 'Scratch Ticket or Draw Game']).agg({'Row ID': 'count','Amount Won': 'sum'}).reset_index()
retailerwinners.columns = ['Retailer Name', 'County', 'Game Type', 'Num Winners', 'Total Prizes']
gametypes = ['All Games'] + sorted(win['Scratch Ticket or Draw Game'].unique().tolist()) # Get all unique games + counties for dropdown menus
counties = ['All Counties'] + sorted(win['Retailer County'].dropna().unique().tolist())

def get_filtered_data(game_type='All Games', county='All Counties'):
    filtered = retailerwinners.copy()
    if game_type != 'All Games':
        filtered = filtered[filtered['Game Type'] == game_type]
    if county != 'All Counties':
        filtered = filtered[filtered['County'] == county]
    result = filtered.groupby('Retailer Name', as_index=False).agg({'Num Winners': 'sum', 'Total Prizes': 'sum'}).sort_values('Num Winners', ascending=False).head(20)
    return result

initial_data = get_filtered_data()
fig = go.Figure()
fig.add_trace(go.Bar(y=initial_data['Retailer Name'], x=initial_data['Num Winners'], orientation='h', marker=dict(color='forestgreen'), text=initial_data['Num Winners'], textposition='auto', hovertemplate='<b>%{y}</b><br>Winners: %{x}<br>Total Prizes: $%{customdata:,.0f}<extra></extra>', customdata=initial_data['Total Prizes']))

gamebuttons = [] # Make dropdown menu for game type
for game in gametypes:
    filtered = get_filtered_data(game_type=game)
    gamebuttons.append(
        dict(label=game, method='update', args=[{'y': [filtered['Retailer Name']], 'x': [filtered['Num Winners']],'text': [filtered['Num Winners']], 'customdata': [filtered['Total Prizes']]}, {'title': f'Top 20 Retailers with Most Lottery Winners in: {game}'}]))
countybuttons = [] # Make dropdown menu for county
for county in counties:
    filtered = get_filtered_data(county=county)
    countybuttons.append(
        dict(label=county, method='update', args=[{'y': [filtered['Retailer Name']], 'x': [filtered['Num Winners']], 'text': [filtered['Num Winners']], 'customdata': [filtered['Total Prizes']]}, {'title': f'Top 20 Retailers with Most Lottery Winners in {county}'}]))

fig.update_layout(
    title='Top 20 Retailers with Most Lottery Winners', xaxis=dict(title='Number of Winners', showgrid=True, gridcolor='lightgray'),
    yaxis=dict(title='', autorange='reversed'), plot_bgcolor='white', paper_bgcolor='white', height=700, margin=dict(l=250, r=100, t=120, b=80),
    font=dict(family="Calibri", size=16, color="black"),
    updatemenus=[
        dict(buttons=gamebuttons, direction='down', showactive=True, x=0.37, xanchor='left', y=1.15, yanchor='top', bgcolor='white', bordercolor='gray', borderwidth=1),
        dict(buttons=countybuttons, direction='down', showactive=True, x=0.7, xanchor='left', y=1.15, yanchor='top', bgcolor='white', bordercolor='gray', borderwidth=1)],
    annotations=[
        dict(text='Filter by Game:', x=0.3, xref='paper', y=1.14, yref='paper', showarrow=False, xanchor='left', font=dict(size=12)),
        dict(text='Filter by County:', x=0.63, xref='paper', y=1.14, yref='paper', showarrow=False, xanchor='left', font=dict(size=12))]
    )

fig.show()
```

The bar chart shows the retailers in Texas with the most lottery winners. It can be filtered by game and county. According to the chart, *Colonies Food Mart* had the most winners (6,467), followed by *Winners Corner TX LLC* (3,990) and *Players Cafe* (1,212). When the bar chart is filtered to show top retailers in Harris County (the county with the greatest number of lottery winners) for scratch ticket winners (the most popular game type), *Julina Food Store*, *Dean's Grocerette INC*, and *Texan Market* have the highest numbers of winners, with 381, 237, and 181, respectively.

### **Research Question #5:** How do lottery claim patterns vary by day of the week and week of the month in 2024?

```python
claimspattern = win.groupby(['Day of Week', 'Week of Month']).size().reset_index(name='Num Claims')
dayorder = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekorder = [1, 2, 3, 4, 5]

claimspattern = claimspattern.set_index(['Day of Week', 'Week of Month']).reindex(pd.MultiIndex.from_product([dayorder, weekorder], names=['Day of Week', 'Week of Month']), fill_value=0).reset_index() # Used Gemini to help
heatmap_data = claimspattern.pivot(index='Day of Week', columns='Week of Month', values='Num Claims').loc[dayorder, weekorder]

fig = px.imshow(
    heatmap_data, labels=dict(x="Week of Month", y="Day of Week", color="Number of Claims"),
    text_auto=True,color_continuous_scale='Greens', title='Claim Patterns by Day of Week and Week of Month (2024)')

fig.update_layout(
    xaxis=dict(tickmode='linear'), yaxis=dict(categoryorder='array', categoryarray=dayorder), plot_bgcolor='white',
    paper_bgcolor='white', height=500, margin=dict(l=120, r=150, t=100, b=80), font=dict(family="Calibri", size=16, color="black"))

fig.update_traces(textfont_size=15)
fig.show()
```

The heatmap shows lottery claim patterns by day of the week and week of the month in 2024. Mondays consistently have the most claims, especially in the first week of the month, while weekends have no claims at all because lottery claims are only processed on business days and claim center offices are closed on weekends. The fifth week of the month also has lower claim counts, which is expected since not all months have a fifth week.
